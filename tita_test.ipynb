```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import pandas as pd

titan_d = './input/titanic/train.csv'
test_d = './input/titanic/test.csv'

titan_data = pd.read_csv(titan_d, index_col='PassengerId')
test_data = pd.read_csv(test_d, index_col='PassengerId')

titan_data.head()
```

```python
# Remove rows with missing target, separate target from predictors
titan_data.dropna(axis=0, subset=['Survived'], inplace=True)
y = titan_data.Survived
titan_data.drop(['Survived'], axis=1, inplace=True)

X = titan_data
X_test = test_data

# To keep things simple, we can choose to use only numerical predictors
#X = X_full.select_dtypes(exclude=['object'])
#X_test = X_test_full.select_dtypes(exclude=['object'])

# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```
